```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")
```

```python
### Distribution
#The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.
#This assumes 12 working months and 50 working weeks.
#Plot the distribution curve for the column `ConvertedComp`.

# Plot the distribution curve for 'ConvertedComp'
plt.figure(figsize=(10, 6))
sns.histplot(df['ConvertedComp'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.show()

```

```python
#What is the median of the column `ConvertedComp`?
median_converted_comp = df['ConvertedComp'].median()

# Print the median of the 'ConvertedComp' column
print("Median of ConvertedComp:", median_converted_comp)
```

```python
#Find out the  median ConvertedComp of responders identified themselves only as a **Woman**?

median_comp_woman = df[df['Gender'] == 'Woman']['ConvertedComp'].median()

# Print the median 'ConvertedComp' for responders identifying as 'Woman'
print("Median ConvertedComp for Responders Identifying as 'Woman':", median_comp_woman)
```

```python
#Give the five number summary for the column `Age`?
five_num_summary_age = df['Age'].describe(percentiles=[.25, .5, .75])

# Print the five-number summary for the 'Age' column
print("Five-Number Summary for 'Age' column:")
print(five_num_summary_age[['min', '25%', '50%', '75%', 'max']])
```

```python
#Plot a histogram of the column `Age`.
# Plot a histogram for the 'Age' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Outliers
#Find out if outliers exist in the column `ConvertedComp` using a box plot?
# Create a box plot for the 'ConvertedComp' column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['ConvertedComp'], color='skyblue')
plt.title('Box Plot of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.show()
```

```python
#Find out the Inter Quartile Range for the column `ConvertedComp`.
# Calculate the Interquartile Range (IQR) for 'ConvertedComp'
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Print the Interquartile Range (IQR)
print("Interquartile Range (IQR) for 'ConvertedComp':", IQR)
```

```python
#Find out the upper and lower bounds.
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Calculate upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Print the upper and lower bounds
print("Lower Bound for Potential Outliers:", lower_bound)
print("Upper Bound for Potential Outliers:", upper_bound)
```

```python
#Create a new dataframe by removing the outliers from the `ConvertedComp` column.
# Calculate upper and lower bounds
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new DataFrame without outliers
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

# Print the shape of the new DataFrame without outliers
print("Shape of DataFrame after removing outliers:", df_no_outliers.shape)
```

```python
#What is the median ConvertedComp after removing outliers?
# Calculate upper and lower bounds for outliers
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new DataFrame without outliers
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

# Calculate the median 'ConvertedComp' after removing outliers
median_comp_after_outliers = df_no_outliers['ConvertedComp'].median()

# Print the median 'ConvertedComp' after removing outliers
print("Median ConvertedComp after Removing Outliers:", median_comp_after_outliers)
```

```python
#Based on the boxplot of ‘Age’ how many outliers do you see below Q1?
# Create a box plot for the 'Age' column
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Age'], color='skyblue')
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()

# Calculate Q1 and IQR for the 'Age' column
Q1_age = df['Age'].quantile(0.25)
IQR_age = df['Age'].quantile(0.75) - Q1_age

# Calculate the lower bound for outliers
lower_bound_age = Q1_age - 1.5 * IQR_age

# Count the number of outliers below Q1
outliers_below_Q1_age = df[df['Age'] < lower_bound_age].shape[0]

# Print the number of outliers below Q1
print("Number of Outliers Below Q1 for Age:", outliers_below_Q1_age)
```

```python
#What is the mean ConvertedComp after removing outliers? 
# Calculate upper and lower bounds for outliers
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new DataFrame without outliers
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

# Calculate the mean 'ConvertedComp' after removing outliers
mean_comp_after_outliers = df_no_outliers['ConvertedComp'].mean()

# Print the mean 'ConvertedComp' after removing outliers
print("Mean ConvertedComp after Removing Outliers:", mean_comp_after_outliers)
```
